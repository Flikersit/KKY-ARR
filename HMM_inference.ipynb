```python
CONST_ND = 1/(np.sqrt(np.power(2*np.pi, 13)))

def calculate_b(m, C, o):
    det = 1/np.sqrt((np.linalg.det(C)))
    return CONST_ND*det*np.exp(-0.5*(np.transpose((o - m))@np.linalg.inv(C)@(o-m)))


path_to_o = r"D:\prednasky\ARR\test_2.txt"
data = np.loadtxt(path_to_o)

b_ano = np.zeros((33, 5), dtype=float)
b_ne = np.zeros((33, 4), dtype=float)


for i in range(data.shape[0]):
    b_ano[i, 1] = calculate_b(
        HMM_ANO["means"][0, :],
        np.diag(HMM_ANO["varis"][0, :]),
        data[i, :]
    )

    b_ano[i, 2] = calculate_b(
        HMM_ANO["means"][1, :],
        np.diag(HMM_ANO["varis"][1, :]),
        data[i, :]
    )

    b_ano[i, 3] = calculate_b(
        HMM_ANO["means"][2, :],
        np.diag(HMM_ANO["varis"][2, :]),
        data[i, :]
    )

    b_ne[i, 1] = calculate_b(
        HMM_NE["means"][0, :],
        np.diag(HMM_NE["varis"][0, :]),
        data[i, :]
    )

    b_ne[i, 2] = calculate_b(
        HMM_NE["means"][1, :],
        np.diag(HMM_NE["varis"][1, :]),
        data[i, :]
    )

print(HMM_ANO["means"][0, :].shape)
print(HMM_ANO["varis"][0, :].shape)
print(data[0, :].shape)
print(b_ano.shape)
print(b_ano)
```
